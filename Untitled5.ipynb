```python
from manim import *
import numpy as np

# 1. Branch and Bound Animation
class BranchAndBoundScene(Scene):
    def construct(self):
        title = Text("Branch and Bound Visualization", font_size=36).scale(0.8)
        self.play(Write(title))
        self.wait(1)
        self.play(FadeOut(title))

        # Create a simple binary tree structure
        root = Dot(point=ORIGIN, color=BLUE)
        left = Dot(point=LEFT*2 + UP, color=BLUE)
        right = Dot(point=RIGHT*2 + UP, color=BLUE)
        left_left = Dot(point=LEFT*3 + DOWN, color=BLUE)
        left_right = Dot(point=LEFT + DOWN, color=BLUE)
        right_left = Dot(point=RIGHT + DOWN, color=BLUE)
        right_right = Dot(point=RIGHT*3 + DOWN, color=BLUE)

        nodes = VGroup(root, left, right, left_left, left_right, right_left, right_right)

        # Create edges
        edges = VGroup(
            Line(root.get_center(), left.get_center()),
            Line(root.get_center(), right.get_center()),
            Line(left.get_center(), left_left.get_center()),
            Line(left.get_center(), left_right.get_center()),
            Line(right.get_center(), right_left.get_center()),
            Line(right.get_center(), right_right.get_center())
        )
        self.play(Create(nodes), Create(edges))
        self.wait(1)

        # Simulate pruning a branch (prune left-right branch)
        prune_edge = edges[3]
        self.play(prune_edge.animate.set_color(RED), FadeOut(prune_edge))
        self.wait(1)
        self.play(FadeOut(nodes), FadeOut(edges))
        self.wait(1)

# 2. Travelling Salesman Problem (TSP) Animation
class TSPScene(Scene):
    def construct(self):
        title = Text("TSP Visualization", font_size=36).scale(0.8)
        self.play(Write(title))
        self.wait(1)
        self.play(FadeOut(title))

        # Generate random cities
        np.random.seed(42)
        cities = [np.array([x, y, 0]) for x, y in np.random.uniform(-3, 3, (8, 2))]
        city_dots = VGroup(*[Dot(point=city, color=WHITE) for city in cities])
        labels = VGroup(*[Text(str(i), font_size=24).next_to(dot, DOWN) for i, dot in enumerate(city_dots)])
        self.play(Create(city_dots), Write(labels))
        self.wait(1)

        # Draw a closed tour using a fixed order for demonstration
        order = [0, 3, 1, 5, 7, 6, 2, 4, 0]
        tour_lines = VGroup()
        for i in range(len(order) - 1):
            line = Line(cities[order[i]], cities[order[i+1]], color=BLUE)
            tour_lines.add(line)
        self.play(Create(tour_lines))
        self.wait(1)

        # Highlight the tour as optimal
        self.play(tour_lines.animate.set_color(RED))
        self.wait(2)
        self.play(FadeOut(city_dots), FadeOut(labels), FadeOut(tour_lines))
        self.wait(1)

# 3. 0/1 Knapsack Problem Animation
class KnapsackScene(Scene):
    def construct(self):
        title = Text("0/1 Knapsack Visualization", font_size=36).scale(0.8)
        self.play(Write(title))
        self.wait(1)
        self.play(FadeOut(title))

        # Draw a knapsack (rectangle)
        knapsack = Rectangle(width=3, height=4, color=WHITE)
        knapsack.to_edge(RIGHT)
        self.play(Create(knapsack))
        self.wait(1)

        # Create items (represented as circles) with labels indicating value/weight
        items = [
            Dot(point=LEFT*3 + UP*2, color=GREEN),
            Dot(point=LEFT*3 + UP, color=GREEN),
            Dot(point=LEFT*3, color=GREEN),
            Dot(point=LEFT*3 + DOWN, color=GREEN)
        ]
        labels = VGroup(*[Text(f"Item {i}", font_size=24).next_to(item, LEFT) for i, item in enumerate(items)])
        self.play(Create(VGroup(*items)), Write(labels))
        self.wait(1)

        # Animate items moving into the knapsack
        for item in items:
            self.play(item.animate.move_to(knapsack.get_center() + np.random.uniform(-1, 1, 3)))
            self.wait(0.5)
        self.wait(2)
        self.play(FadeOut(knapsack), FadeOut(VGroup(*items), labels))
        self.wait(1)

# 4. Suffix Arrays and Suffix Trees Animation
class SuffixStructuresScene(Scene):
    def construct(self):
        title = Text("Suffix Arrays & Trees Visualization", font_size=36).scale(0.8)
        self.play(Write(title))
        self.wait(1)
        self.play(FadeOut(title))

        # Display a sample string
        sample_string = "BANANA"
        string_display = Text(f"String: {sample_string}", font_size=36)
        self.play(Write(string_display))
        self.wait(1)

        # Generate and display suffixes
        suffixes = [sample_string[i:] for i in range(len(sample_string))]
        suffix_group = VGroup(*[Text(s, font_size=24) for s in suffixes]).arrange(DOWN, aligned_edge=LEFT).next_to(string_display, DOWN, buff=1)
        self.play(Write(suffix_group))
        self.wait(1)

        # Simulate sorting of suffixes by transforming the group to sorted order
        sorted_suffixes = sorted(suffixes)
        sorted_group = VGroup(*[Text(s, font_size=24) for s in sorted_suffixes]).arrange(DOWN, aligned_edge=LEFT).next_to(string_display, RIGHT, buff=1)
        self.play(Transform(suffix_group, sorted_group))
        self.wait(2)
        self.play(FadeOut(string_display), FadeOut(suffix_group))
        self.wait(1)

# 5. NP-Complete and NP-Hard Problems Animation
class NPCompleteScene(Scene):
    def construct(self):
        title = Text("NP-Complete & NP-Hard Problems", font_size=36).scale(0.8)
        self.play(Write(title))
        self.wait(1)
        self.play(FadeOut(title))

        # Create a maze-like grid to represent computational complexity
        maze_lines = VGroup(
            *[Line(LEFT*3 + UP*i, RIGHT*3 + UP*i, color=WHITE) for i in range(-2,3)],
            *[Line(UP*2 + RIGHT*i, DOWN*2 + RIGHT*i, color=WHITE) for i in range(-3,4)]
        )
        self.play(Create(maze_lines))
        self.wait(1)

        # Overlay text representing NP-Completeness
        np_text = Text("NP-Complete", font_size=36, color=RED)
        np_text.move_to(ORIGIN)
        self.play(Write(np_text))
        self.wait(2)
        self.play(FadeOut(maze_lines), FadeOut(np_text))
        self.wait(1)
```
